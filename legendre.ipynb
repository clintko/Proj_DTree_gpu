# Legendre Polynomials

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.special import legendre

# print(plt.style.available)
plt.style.use("seaborn-whitegrid")
```

```python
N = 1000
xvals = np.linspace(-1, 1, N)
```

```python
def Legendre(x, n):
    """Set up a function defining Legendre Polynomials"""
    # calling the n_th deg of Legendre polynomials
    leg = legendre(n)
    
    # 
    y = leg(x)
    return y
```

```python
yvals_leg1 = Legendre(xvals, 1)
yvals_leg2 = Legendre(xvals, 2)
yvals_leg3 = Legendre(xvals, 3)
```

```python
roots1, weights1 = np.polynomial.legendre.leggauss(1)
roots2, weights2 = np.polynomial.legendre.leggauss(2)
roots3, weights3 = np.polynomial.legendre.leggauss(3)
```

```python
w1 = np.polynomial.legendre.legweight(roots1)
w2 = np.polynomial.legendre.legweight(roots2)
w3 = np.polynomial.legendre.legweight(roots3)
```

```python
print(weights1)
print(w1)

print(weights2)
print(w2)

print(weights3)
print(w3)
```

```python
plt.plot(xvals, yvals_leg1, 'r--', label = "n = 1")
plt.plot(xvals, yvals_leg2, 'b--', label = "n = 2")
plt.plot(xvals, yvals_leg3, 'g--', label = "n = 3")

plt.plot(roots1, np.repeat(0, 1), 'ro')
plt.plot(roots2, np.repeat(0, 2), 'bo')
plt.plot(roots3, np.repeat(0, 3), 'go')

plt.title("First 3 P_n")
plt.legend()
plt.show()
```

How to put the legend out of the plot

```python
plt.figure(figsize=(10,5))    

for deg in range(1, 16):
    yvals = Legendre(xvals, deg)
    plt.plot(xvals, yvals, label = "n = " + str(deg))


plt.title("P_n")
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
plt.show()
```
